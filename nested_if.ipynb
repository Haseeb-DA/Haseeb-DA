```r

# Taking if raining, temperature and wind from user and suggesting activities
temperature_input <- readline(prompt= " Please enter temperature for the day in celsius : ")

temperature = as.numeric(temperature_input)
print(temperature_input)


wind_input <- readline(prompt= " Please enter wind speed for the day in km/hr : ")
wind = as.numeric(wind_input)
print(wind)

rain_input <-readline(prompt= "Please tell , if its raining TRUE/FALSE ")
rain = as.logical(rain_input)
print(rain)
print(class(rain))



# Suggesting activities for user based on temperature and Wind
# condition 1: temperature >= 40 , wind < 10 - high temperature and less wind, Pool/beach activities
# condition 2: temperature >= 40 , wind >10 - high temperature and more wind, prefer indoor activities
# condition 3: temperature >= 20 and < 40  - Moderate summer weather , less windy - Prefer Cycling/Hiking,more windy - wear Jacket and have Hiking
# condition 4: temperature < 20 Cold weather ,less windy - Prefer Indoor with a bit of walk ,more windy - Prefer Indoor


#Rain No - Existing code

#if it rains, we will decide  by temperature and wind

if(!rain){#If it doesn't rain
if(temperature>=40){
  if(wind<=10){
    print("Very Hot Sunny day and less windy :Prefer Pool/beach activities")
  }else{
    print("Very Hot Sunny day and windy :Prefer indoors")
  }

}else if(temperature>=20){
  if(wind<=10){
    print("Moderate summer weather and less windy :Prefer Cycling/Hiking")
  }else{
    print("Moderate summer weather and windy :Wear Jacket and enjoy Hiking ")
  }
}else{
  if(wind<=10){
    print("cold weather and less windy :Prefer Indoor with a bit of walk")
  }else{
    print("cold weather and windy :Prefer indoor activities")
  }
}


}else{

#It rains, drilling down by temperature and wind

#Condition 1: If rains and windy , irespective of temperature, prefer indoor activity
#Condition 2: If rains and not windy , umbrella or raincoat depending on temperature


if(temperature>=40){
  if(wind<=10){
    print("Very Hot Sunny day and less windy but rainy :Prefer Pool/beach activities with umbrella/raincoat")
  }else{
    print("Very Hot Sunny day and windy and rainy :Prefer indoors")
  }

}else if(temperature>=20){
  if(wind<=10){
    print("Moderate summer weather and less windy but rainy :Prefer Cycling/Hiking with a raincoat")
  }else{
    print("Moderate summer weather and windy and rainy :Wear Jacket and enjoy Hiking with a raincoat")
  }
}else{
  if(wind<=10){
    print("cold weather and less windy but rainy :Prefer Indoor activities")
  }else{
    print("cold weather and windy and rainy:Prefer indoor activities")
  }
}


}#end of else rain clause
```
